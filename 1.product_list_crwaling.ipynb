# 선제품별 링크들을 크롤링하는 코드입니다.

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import random
import csv
import re
```

#### olive url(인기순과 구매순을 나누어서 인기순은 댓글 크롤링, 구매순은 따로 구매순위만 크롤링하였습니다)

```python
url = "https://www.oliveyoung.co.kr/store/display/getMCategoryList.do?dispCatNo=100000100110006&fltDispCatNo=&prdSort=01&pageIdx={}&rowsPerPage=24&searchTypeSort=btn_thumb&plusButtonFlag=N&isLoginCnt=0&aShowCnt=0&bShowCnt=0&cShowCnt=0&trackingCd=Cat100000100110006_Small&amplitudePageGubun=&t_page=&t_click=&midCategory=%EC%84%A0%ED%81%AC%EB%A6%BC&smallCategory=%EC%A0%84%EC%B2%B4&checkBrnds=&lastChkBrnd="
```

```python
driver = webdriver.Chrome()
#suncream_link.
with open('./crawling_data/suncream_link.csv', "w", newline="", encoding = "utf-8-sig") as fw:
    writer = csv.DictWriter(fw, fieldnames=['page', 'brand_name', 'product_name', 'product_link', 'image_url'])# 헤더 이름 지정
    writer.writeheader()
    for i in range(1, 17):
        new_url = url.format(i)
        driver.get(new_url)
        timesleep = random.randint(1, 5)  
        time.sleep(timesleep)
        goods = driver.find_elements(By.CLASS_NAME, 'prd_info ')
        for good in goods:
            try:
                product_link = good.find_element(By.TAG_NAME, 'a')# 제품링크
                brand_name = good.find_element(By.CLASS_NAME, 'tx_brand').text# 브랜드명
                product_name = good.find_element(By.CLASS_NAME, 'tx_name').text# 제품명
                img = good.find_element(By.TAG_NAME, 'img')# 이미지
                imgsrc = img.get_attribute('src')
                writer.writerow({
                    'page': i,# 크롤링 화면 페이지
                    'brand_name': brand_name,#선제품 브랜드
                    'product_name': product_name,#선제품명
                    'product_link': product_link.get_attribute('href'),#해당 제품 링크
                    'image_url': imgsrc#image url
                })
                print(f"brand: {brand_name}, Link: {product_link}")
            except:
               
                print("a 태그를 찾을 수 없음")
                continue   
driver.quit()
```
